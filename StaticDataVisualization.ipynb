# 2.  Generate a plot that details the number of Beer, Wine and Spirit Servings for each of the 5 top Countries

### 2.1 - Plot of number of beer per top five countries

In order to generate a plot that shows the number of drinks servings for each of the top 5 countries, I will filter each of the top 5 countries for each type of drinks then generate a barplot to show the distribution for each of the top countries per drinks. For instance, first I will filter the data using a loc attribute and filter the data into a dataframe and slice it. The first barplot will be the top 5 countries of beer_servings, then the top 5 countries of spirit_serving and wine servings.

I will plot it using bar plot because it is useful to describe relationships between categorical and numerical variable. So this barplot provide with feature to plot the distribution of values between the total number of beer and countries from pandas Data Frame. 

Sort beer_servings values in descending order to cheek the country which consume the most amount of beer.

```python
df.sort_values('beer_servings', ascending=False)
```

### Filtering the data

I have found that the top 5 countries of beer_servings is from 346 servings to 376
So I create a condition to filter the data, using a loc attribute 
where the servings are >= 343 and store in a different 
data frame called df_total_beer

```python
df_total_beer = df.loc[df.beer_servings >= 343]
df_total_beer
```

Afer using the condition I filtered the df_total_beer dataframe to get only the 5 rows, using the nlargest() funtion to return the fisrt 5 rows

```python
df_total_beer_filter = df_total_beer.nlargest(5, 'beer_servings' )
df_total_beer_filter
```

Slicing the data, selecting the countries and beer_servings only.

```python
df_total_beer_filter[['country', 'beer_servings']]
```

Libraries for ploting

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from numpy import median, mean
```

Creating a barplot using seaborn to show the number of beer of the top 5 countries.

```python
ax = sns.barplot(x = 'country' , y = 'beer_servings', data = df_total_beer_filter, palette="Blues_d")
#modifing x and y-axis tick labels
ax.set_title("Total number of Beer per Top 5 Countries", fontdict={'fontsize' : 16})
ax.set_xlabel('Country', fontdict={'fontsize' : 15})
ax.set_ylabel('Total Number of Beer', fontdict={'fontsize' : 15})
#modifying the font-size and rotation of x-axis and set fontsize and rotation of a-axis tick labels
ax.set_xticklabels(ax.get_xticklabels(), fontsize=13, rotation=30)
```

For this visualization I will use a violin plot from seaborn, because it is useful to show categorical variables and its relationship with numerical variables. It shows the distribution of the total litres of alcohol by continent and the density in the distribution and its useful to create the plot from pandas dataframe as the names are used to interpret the axis.

## Creating a barplot using seaborn to show the continent with the most alsohool consumption.

```python
ax = sns.barplot(x = 'continent' , y = 'total_litres_of_pure_alcohol', data = df_total_alcohol, palette="Blues_d", estimator = mean, 
                 ci=68, order=['Europe', 'South America', 'North America','Oceania', 'Africa', 'Asia'])
#set title)
#modifing x and y-axis tick labels

ax.set_title("Continent with Most Alcohol Consumption", fontdict={'fontsize' : 16})
ax.set_xlabel('Continent', fontdict={'fontsize' : 15})
ax.set_ylabel('Alcohol Consumption', fontdict={'fontsize' : 15})
#modifying the font-size and rotation of x-axis and set fontsize and rotation of a-axis tick labels
ax.set_xticklabels(ax.get_xticklabels(), fontsize=13, rotation=30)
```

# 5.  Observe the average beer and wine servings by continent

For this visualization I will use barplot using seaborn, in order to show mean average beer and wine serving by continent. Bar graphs are very useful for when there is categorical and numerical variable, and in this case, to show the average distribution with the estimator = mean.

```python
df.head()
```

pd.crosstab() funtion building a cross-tabulation table for beer and wine servings for continent and subtotal with (margins).

```python
pd.crosstab([df.beer_servings, df.wine_servings], df.continent,margins=True, margins_name="Total")
```

Average distribution of beer and wine by continent.

```python
df_average_b_w = pd.crosstab([df.beer_servings, df.wine_servings], df.continent)
df_average_b_w.mean(axis = 0)
```

## Barplot of Average Beer Serving per continent

The same observation mean can be seen in the graph using seaborn, a mean distribution of beer_servings per continent.

```python
sns.set(style = 'whitegrid')
#using ci=68 parameter to set it to 68% and using order to reorder the x axis
ax = sns.barplot(x = 'continent', y = 'beer_servings', data = df, palette="Blues_d", estimator = mean, 
                 ci=68, order=['Europe', 'South America', 'North America','Oceania', 'Africa', 'Asia'])
#set title
ax.set_title("Average Beer Serving per Continent", fontdict={'fontsize' : 16})
#modifing x and y-axis tick labels
ax.set_xlabel('Continent', fontdict={'fontsize' : 15})
ax.set_ylabel('Beer Servings', fontdict={'fontsize' : 15})
#modifying the font-size and rotation of x-axis and set fontsize and rotation of a-axis tick labels
ax.set_xticklabels(ax.get_xticklabels(), fontsize=13, rotation=30)
```
